This next section reads in data regariding the use of contraceptives to prevent pregnancies and abortions. The data is manipulated to separate teh abortions averted and the pregnancies averted. 

```python
contraceptives_data = pd.read_csv("contraception_data.csv")
```

```python
contraceptives_data.dropna(subset = ["datum"])

abortions_averted = contraceptives_data[0:51].copy()

pregnancies_averted = contraceptives_data[103:153]

abortions_averted["datum"] = abortions_averted["datum"].apply(pd.to_numeric)

abortions_averted_state = abortions_averted["state_id"]
abortions_averted_rate = abortions_averted["datum"]

pregnancies_averted_state = pregnancies_averted["state_id"]
pregnancies_averted_rate = pregnancies_averted["datum"]

```

After aggregating the data, the graph below is created by establishing a figure and then printing a bar graph of the abortions that were averted using contraceptives (by public clinics). The data is displated for all 50 states.

```python
plt.figure(figsize = (20,7))
plt.bar(abortions_averted_state, abortions_averted_rate,  width=0.5, align='center')
plt.title('Abortions Averted Using Contraceptives in all 50 states')
plt.xlabel('State ID')
plt.ylabel('Number of Abortions Averted')
plt.show()
```

This next graph shows the pregnancies that were averted in all 50 states by using contraceptives.

```python
plt.figure(figsize = (20,7))
plt.bar(pregnancies_averted_state, pregnancies_averted_rate,  width=0.5, align='center')
plt.title('Pregnancies Averted Using Contraceptives in all 50 states')
plt.xlabel('State ID')
plt.ylabel('Number of Pregnancies Averted')
plt.show()
```
